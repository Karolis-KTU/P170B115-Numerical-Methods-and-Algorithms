```python
import numpy as np
import matplotlib.pyplot as plt

# given function
def fx(x):
  return -0.3*x**5 - 1.10*x**4 + 1.14*x**3 + 3.84*x**2 + 1.48*x - 0.22

# initial data
xmin, xmax = -4.67, 3.34
h = 0.2

# ploting given function
dx = 0.1;  x=np.arange(xmin,xmax, dx); y=fx(x)
plt.xlabel("x"); plt.ylabel("y"); plt.plot(x, y, 'b'); plt.grid()

# bracketing roots
while xmin<xmax:
  if np.sign(fx(xmin)) !=np.sign(fx(xmin+h)):
    print("Root : [%.3f ;  %.3f]" % (xmin,xmin+h))
    plt.plot([xmin], [0], 'or')
    plt.plot([xmin+h], [0 ], 'og')
  xmin += h
```

Pusiaukirtos (Bisekcijos) metodas

```python
import numpy as np
import matplotlib.pyplot as plot
from IPython.display import display, clear_output

# Lygties funkcija
def LF(x):
    s = 0.3 * x**5 + 1.10 * x**4 - 1.14 * x**3 - 3.84 * x**2 - 1.48 * x + 0.22
    #s = 2 - np.log(x) * np.sin(x**2)
    return s

# Programa
delay = 0.5           # pauses ilgumas grafiniam vaizdavimui
eps = 0.00001            # reikalaujamas tikslumas
xrange = [-4.67,3.34]    # LF grafiko vaizdavimo sritis
#xrange = [6,9]
x1 = -4.07; x2 = -3.84      # atskyrimo intervalas arba du artiniai
maxiter=30          # didziausias leistinas iteraciju skaicius

f1 = LF(x1)
f2 = LF(x2)

# Grafika
X = np.linspace(xrange[0], xrange[1], 500)  # funkcijos vaizdavimo taskai
fig = plot.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('x')
ax.set_ylabel('y')
plot.ylim(-50,50)
#plot.xlim(1,4)
c1, = ax.plot(X, LF(X), 'b')
c2, = ax.plot(xrange, [0, 0], 'k.--')
plot.draw()
plot.pause(delay)
m1, = ax.plot([x1, x1], [0, f1], 'mo--')
plot.draw()
m2, = ax.plot([x2, x2], [0, f2], 'co--')
plot.draw()

display(fig)
#plot.draw()
#plot.pause(delay)

# Iteracijos
for i in range(1, maxiter + 1):  # saknies tikslinimo iteracijos
    xmid = (x1 + x2) / 2  # intervalo vidurys
    fmid = LF(xmid)

    # Grafika
    mmid, = ax.plot([xmid, xmid], [0, fmid], 'ks--')
    plot.draw()
    #plot.pause(delay)

    if np.sign(f1) != np.sign(fmid):
        x2 = xmid
        f2 = fmid
    else:
        x1 = xmid
        f1 = fmid

    # Grafika
    if np.sign(f1) != np.sign(fmid):
        m2.remove()
        mmid.remove()
        m2, = ax.plot([x2, x2], [0, f2], 'co--')
    else:
        m1.remove()
        mmid.remove()
        m1, = ax.plot([x1, x1], [0, f1], 'mo--')
    #clear_output(wait=True)
    display(fig)
    #plot.draw()
    #plot.pause(delay)
    
    # Tikslumas
    tksl = abs(x1 - x2)
    str = "%2d.   x1= %-20g   x2= %-20g  f1= %-20g   f2= %-20g  tikslumas= %-20g" % (i, x1, x2, f1, f2, tksl)
    print(str)
    if tksl < eps:
        str = "\nIsspresta: x=  %-25g  f=   %-25g   tikslumas=%-25g\n" % (x1, f1, tksl)
        break
    elif i == maxiter:
        str = "\nTikslumas nepasiektas"

    
print(str)
```

Tikslinimas naudojant Kvazi-Niutono metodą

```python
import numpy as np
import matplotlib.pyplot as plot
from IPython.display import display, clear_output

# Lygties funkcija
def LF(x):
    s = 0.3 * x**5 + 1.10 * x**4 - 1.14 * x**3 - 3.84 * x**2 - 1.48 * x + 0.22
    #s = 2 - np.log(x) * np.sin(x**2)
    return s

# Programa
delay = 0.5           # pauses ilgumas grafiniam vaizdavimui
eps = 1e-8            # reikalaujamas tikslumas
xrange = [-4.67,3.34]    # LF grafiko vaizdavimo sritis
#xrange = [6,9]
x1 = 1.86; x2 = 2.06      # atskyrimo intervalas arba du artiniai
maxiter=30          # didziausias leistinas iteraciju skaicius

f1 = LF(x1)
f2 = LF(x2)

# Grafika
X = np.linspace(xrange[0], xrange[1], 500)  # funkcijos vaizdavimo taskai
fig = plot.figure()
ax = fig.add_subplot(1, 1, 1)
plot.ylim(-25,25)
#plot.xlim(1,4)
ax.set_xlabel('x')
ax.set_ylabel('y')
c1, = ax.plot(X, LF(X), 'b')
c2, = ax.plot(xrange, [0, 0], 'k.--')
plot.draw()
plot.pause(delay)
m1, = ax.plot([x1, x1], [0, f1], 'mo--')
plot.draw()
m2, = ax.plot([x2, x2], [0, f2], 'co--')
plot.draw()

st, = ax.plot([x2, x1], [f2, f1], 'k.-.')

display(fig)
# plot.draw()
# plot.pause(delay)

# Iteracijos
for i in range(1, maxiter + 1):  # saknies tikslinimo iteracijos

    xmid = x1 - (x1 - x2) / (f1 - f2) * f1  # kvazi-liestines nulio taskas
    fmid = LF(xmid)

    # Grafika
    mmid, = ax.plot([xmid, xmid], [0, fmid], 'ks--')
    st.remove()
    st, = ax.plot([x2, x1, xmid], [f2, f1, 0], 'k.-.')



    plot.draw()
    #plot.pause(delay)

    x2 = x1
    f2 = f1
    x1 = xmid
    f1 = fmid

    # Grafika
    mmid.remove()
    m1.remove()
    m2.remove()
    m1, = ax.plot([x1, x1], [0, f1], 'mo--')
    m2, = ax.plot([x2, x2], [0, f2], 'co--')

    #clear_output(wait=True)
    display(fig)

    #plot.draw()
    #plot.pause(delay)

    # Tikslumas
    tksl = abs(x1 - x2)
    str = "%2d.   x1= %-20g   x2= %-20g  f1= %-20g   f2= %-20g  tikslumas= %-20g" % (i, x1, x2, f1, f2, tksl)
    print(str)
    if tksl < eps:
        str = "\nIsspresta: x=  %-25g  f=   %-25g   tikslumas=%-25g\n" % (x1, f1, tksl)
        break
    elif i == maxiter:
        str = "\nTikslumas nepasiektas"

print(str)
```

```python
import matplotlib.pyplot as plt
import sympy as sp
import numpy as np
import math

MAX_ERROR = 1e-4 # Maximum acceptable error

def h(x):
    return -77 * sp.sin(x) + 25 - 3 * x

def plot_draw(x_range, taylor, n_value):
    plt.plot(x_range, npoints_range, color="red", linestyle="solid")
    plt.axhline(0, color = "black", linestyle="dashed")

    plt.xlabel("x")
    plt.ylabel("h(x), TE(x)")
    plt.grid()

    plt.plot(x_range, taylor, color="blue", linestyle="dashed")
    plt.legend(['h(x)','Ox', f'TE = {n_value}'])

def Taylor(h, x, x0, n):
    hp = h.subs(x, x0)
    dh = h
    for i in range(1, n + 1):
        dh = dh.diff(x)
        hp = hp + dh.subs(x, x0) * (x - x0)**i / math.factorial(i)

    return hp

def function_roots(start, end, interval, x_symbol, h_symbol):
    root_intervals = scan(start, end, interval, x_symbol, h_symbol)
    output = []
    for i in range(len(root_intervals)):
        root_value = quasi_newton_method(root_intervals[i][1], 35, x_symbol, h_symbol)
        output.append((root_intervals[i][0], root_intervals[i][1], root_value))
    return output

def quasi_newton_method(x_value, max_iteration_count, x_symbol, h_symbol):
    x1 = x_value
    x2 = x1 + 0.01  # Initialize x2 as a small increment from x1
    f1 = h_symbol.subs(x_symbol, x1)
    f2 = h_symbol.subs(x_symbol, x2)

    for i in range(1, max_iteration_count + 1):
        xmid = x1 - (x1 - x2) / (f1 - f2) * f1
        fmid = h_symbol.subs(x_symbol, xmid)

        x2 = x1
        f2 = f1
        x1 = xmid
        f1 = fmid

        tksl = abs(x1 - x2)

        if tksl < MAX_ERROR:
            return x1

    return None
        
def scan(start, end, interval, x_symbol, h_symbol):
    output = []
    x = start
    while x < end:
        if np.sign(h_symbol.subs(x_symbol, x)) != np.sign(h_symbol.subs(x_symbol, x + interval)):
            output.append((x, x + interval))
        x += interval
    return output

def get_error_rate(x1, x2):
    return np.abs(x1 - x2)

# Main
x, hp = sp.symbols('x hp')
xmin = -4
xmax = 6
npoints = 400
x0 = 0 # Point of expansion
N = 50 # Max number of terms
x_range = np.linspace(xmin, xmax, npoints)
h_roots = function_roots(xmin, xmax, 0.1, x, h(x))

npoints_range = []
for i in range(npoints):
    npoints_range.append(h(x).subs(x, x_range[i]))

taylor_roots_array = []

# Find the correct taylor series
is_taylor_found = False
for n in range(1, N + 1):
    # Plot the graph
    taylor_values = []
    for i in range(npoints):
        taylor_values.append(hp.subs(x, x_range[i]))

    # Check if taylor series is found
    if(is_taylor_found):
        print(f"Taylor series found N={n-1}")
        plot_draw(x_range, taylor_values, n-1)
        break

    # Taylor series calculation
    hp = Taylor(h(x), x, x0, n)

    # get taylor roots
    taylor_roots = function_roots(xmin, xmax, 0.1, x, hp)
    taylor_roots_array.append(taylor_roots)
    #print(f"Taylor roots: {taylor_roots}")

    # Check if the taylor series error is small enough
    is_current_taylor_found = True
    for i in range(len(taylor_roots)):
        h_root = h_roots[i][2]
        taylor_root = taylor_roots[i][2]
        error_rate = get_error_rate(h_root, taylor_root)
        if (error_rate > MAX_ERROR):
            is_current_taylor_found = False
            break

    if(len(taylor_roots) == len(h_roots) and is_current_taylor_found):
        is_taylor_found = True

ans = sp.Poly(hp) 
plt.show()
print("Grafikas, kuris nurodo visą randamų šaknų skaičių nagrinėjamame intervale")
plt.clf()
plt.xlabel("TE eilė")
plt.ylabel("Šaknų skaičius")
plt.plot(range(1, len(taylor_roots_array)+1), [len(x) for x in taylor_roots_array], color='green', linestyle='solid', marker='o', linewidth=2)
plt.xticks(range(1, len(taylor_roots_array)+1))
plt.show()
```
